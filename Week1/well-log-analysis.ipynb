## Data Cleaning

```python
def dataClean(filepath):
    dataframe = pd.read_csv(f"{filepath}")
    dataframe.columns = dataframe.iloc[1]
    dataframe =  dataframe[2:]
    dataframe['Date']=pd.to_datetime(dataframe['Date'])
    return dataframe
    
```

## Monthly data production 

## Analysis

```python
def MonthlyAnalysisOfColumns(filepath,Column):
    
    # clean data
    df = dataClean(f"{filepath}")
    # making sure all values in columns are 
    
    df[f"{Column}"] = pd.to_numeric(df[f"{Column}"], errors='coerce')

    # setting the index of the dataframe as the date
    df.set_index('Date', inplace=True)

    # resampling of columns by the mean

    dfColumn = df[f"{Column}"].resample('ME').mean()
    # we would get a series where the date is the column so we reset the index
    dfColumn= dfColumn.reset_index()


    dfColumn['Date'] = dfColumn['Date'].dt.to_period('M').dt.to_timestamp()    



    
    plt.plot(dfColumn['Date'], dfColumn[f"{Column}"], color='blue', linewidth=2)
    plt.title(f"Monthly Average {Column} (Well â„–807)")
    plt.xlabel("Month")
    plt.ylabel(f"{Column}")

    plt.grid(True)
    plt.show()
    
    
```

```python
 MonthlyAnalysisOfColumns('./Oil well.csv',"Water cut (%) ")
```

```python
 MonthlyAnalysisOfColumns('./Oil well.csv',"\nOil volume (m3/day)")
```
